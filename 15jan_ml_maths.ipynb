```python
# prompt: import numpy as np

import numpy as np
```

```python
X = 2* np.random.rand(100,1)
```

```python
#y = mx+c
y = 3*X + 4
```

```python
import matplotlib.pyplot as plt
plt.plot(X,y)
```

```python
plt.scatter(X,y, color='red')
```

```python
w = np.random.randn(1)
b = np.random.randn(1)
```

```python
learning_rate = 0.1
```

```python
m = len(X)
```

```python
loss_history=[]
```

```python
for i in range(100):
  y_pred = w*X + b

  loss = np.mean((y-y_pred)**2)
  loss_history.append(loss)

  dw = -2/m * np.sum(X*(y-y_pred))
  db = -2/m * np.sum((y-y_pred))
  w = w - learning_rate * dw
  b = b- learning_rate*db

  if i%10==0:
    print("loss:", loss, "current weight:", w, "current bias:", b)

    print(f"final parameters (w): {w} and (b): {b}")
```

```python
plt.plot(range(100), loss_history, color="red", label="Loss Curve")
```

```python
plt.scatter(X,y, color="blue", label="Data points")
plt.plot(X,w*X+b, color="red", label="Fitted Line")
```
